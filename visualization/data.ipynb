```python
count_df = df.groupby(['State', 'SmokerStatus']).size().reset_index(name='count')

# Calculate the total counts per state
total_counts = df.groupby('State').size().reset_index(name='total')

# Merge the counts data with the total counts
percentage_df = pd.merge(count_df, total_counts, on='State')

# Calculate the percentage
percentage_df['percentage'] = (percentage_df['count'] / percentage_df['total']) * 100
```
